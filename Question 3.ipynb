```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

## Question 3
df2 = pd.read_excel('Combined.xlsx')
df2.columns = ['Date', 'Number of Daily Cases', 'Moving Average']
bar_graph = df2.plot('Date', ['Moving Average'], xticks = [],kind = 'bar', color = 'red', width = 1)
df2.plot('Date', ['Number of Daily Cases'], xticks = [], color = 'blue', ax = bar_graph, title = 'Question 3', ylabel = 'Count')
plt.show()
```
